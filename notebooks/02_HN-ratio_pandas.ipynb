```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../TSVfile/rice_result_2401.tsv", sep='\t')
display(df)
```

```python
df.describe()
```

```python
HN = pd.read_csv('../HN-pair_rice.csv', sep=',')
display(HN)
```

```python
HNratio = pd.DataFrame()
HNratio["GENEID"] = df["GENEID"] 
for i in range(len(HN)):
    h = str(HN.iloc[i,2])
    n = str(HN.iloc[i,3])
#    c = str(h) +"/"+str(n)
    HNratio[h] = (df.loc[:,h] + 1)/(df.loc[:,n] +1)
    HNratio.append(HNratio)

display(HNratio)
```

```python
HNratio_ = HNratio.set_index('GENEID')

HNratiolog2 = HNratio_.apply(np.log2)
HNratiolog2 = HNratiolog2.reset_index()
display(HNratiolog2)
```
